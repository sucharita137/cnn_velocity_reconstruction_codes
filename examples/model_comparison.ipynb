```python
import sys
import os
import pandas as pd
```

```python
project_root = os.path.abspath("..")
if project_root not in sys.path:
    sys.path.insert(0, project_root)
```

```python
from kszcnn.cnn_vel_density import train_and_eval as run_density
from kszcnn.cnn_vel_density_pos import train_and_eval as run_density_pos
from kszcnn.cnn_vel_density_pos_mass import train_and_eval as run_density_pos_mass
```

```python
results = []

print("Running density-only model...")
res_den = run_density(make_plots=False)
res_den["model"] = "Density only"
results.append(res_den)

print("Running density + position model...")
res_den_pos = run_density_pos(make_plots=False)
res_den_pos["model"] = "Density + position"
results.append(res_den_pos)

print("Running density + position + halo mass model...")
res_den_pos_mass = run_density_pos_mass(make_plots=False)
res_den_pos_mass["model"] = "Density + position + halo mass"
results.append(res_den_pos_mass)
```

```python
df = pd.DataFrame(results)[
    ["model", "rho_cnn", "rho_vlin", "bias", "rmse"]
]

df
```

```python
df_rounded = df.copy()
for c in ["rho_cnn", "rho_vlin", "bias", "rmse"]:
    df_rounded[c] = df_rounded[c].astype(float).round(3)

df_rounded
```

```python
df_rounded.to_csv("cnn_input_comparison.csv", index=False)
print("Saved cnn_input_comparison.csv")
```
